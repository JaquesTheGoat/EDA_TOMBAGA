```python
#check outliers for said column
def find_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    outliers = column[(column < (Q1 - 1.5 * IQR)) | (column > (Q3 + 1.5 * IQR))]
    
    #display outliers in list form
    if not outliers.empty:
        return f"Outliers detected in {column.name}: {', '.join(map(str, outliers.unique()))}"
    else:
        return f"No outliers detected in {column.name}."

#output
released_year_outliers = find_outliers(data['released_year'])
print(released_year_outliers)
```

### **Temporal Trends**

```python
tracks_per_year = data['released_year'].value_counts().sort_index()

# Create a DataFrame for easier plotting
tracks_per_year_data = tracks_per_year.reset_index()
tracks_per_year_data.columns = ['released_year', 'track_count']

# Plotting the number of tracks released per year
plt.figure(figsize=(14, 6))
sns.barplot(x='released_year', y='track_count', data=tracks_per_year_data, palette='viridis')
plt.title("Number of Tracks Released Per Year")
plt.xlabel("Released Year")
plt.ylabel("Number of Tracks")
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.show()
```

```python
# Group by 'released_month' to count the number of tracks released per month
monthly_releases = data['released_month'].value_counts().sort_index()

# Plot the distribution of track releases per month
plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_releases.index, y=monthly_releases.values, palette="viridis")
plt.xlabel("Month")
plt.ylabel("Number of Tracks Released")
plt.title("Number of Tracks Released per Month")
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()

# Identify the month with the most releases
most_releases_month = monthly_releases.idxmax()
most_releases_count = monthly_releases.max()

# Convert month number to month name
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
most_releases_month_name = month_names[most_releases_month - 1]  # Subtract 1 since month is 1-based

print(f"The month with the most releases is {most_releases_month_name} with {most_releases_count} tracks.")

# Summary of pattern based on the descriptive statistics
mean_month = data['released_month'].mean()
median_month = data['released_month'].median()

# Convert mean and median months to month names
mean_month_name = month_names[int(round(mean_month)) - 1]
median_month_name = month_names[int(median_month) - 1]

print(f"On average, most tracks are released around {mean_month_name} (mean), with the median release month being {median_month_name}.")
```

### **Genre and Music Characteristics**

```python
columns_of_interest = ['streams', 'bpm', 'danceability_%', 'energy_%', 'valence_%', 'acousticness_%']
correlation_matrix = data[columns_of_interest].corr()

# Display the correlation matrix as a heatmap with viridis color palette and transparency
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='inferno', vmin=-1, vmax=1, fmt=".2f", linecolor='Black')
plt.title("Correlation between Streams and Musical Attributes")
plt.show()
```

```python
# Calculate correlations between 'streams' and the specified musical attributes
correlationstreams = data[['streams', 'bpm', 'danceability_%', 'energy_%']].corr()['streams'].drop('streams')

# Convert the Series to a DataFrame for a larger, more readable table format
correlation_df = pd.DataFrame(correlationstreams).reset_index()
correlation_df.columns = ['Musical Attribute', 'Correlation with Streams']

# Display the correlation table in a larger format
print("Correlation of Streams with Musical Attributes:\n")
print(correlation_df)
```

```python
# Calculate the correlation between specific pairs of musical attributes
attribute_pairs = data[['danceability_%', 'energy_%', 'valence_%', 'acousticness_%']].corr()

# Extract specific pairs for clarity
danceability_energy_corr = attribute_pairs.loc['danceability_%', 'energy_%']
valence_acousticness_corr = attribute_pairs.loc['valence_%', 'acousticness_%']

# Create a DataFrame to display these correlations in a larger table format
correlation_df = pd.DataFrame({
    'Attribute Pair': ['Danceability vs Energy', 'Valence vs Acousticness'],
    'Correlation': [danceability_energy_corr, valence_acousticness_corr]
})

# Display the organized table of correlations
print("Correlation between Selected Musical Attribute Pairs:\n")
print(correlation_df)
```

### **Report on the Correlation between Streams and Musical Attributes**
#### **Heatmap Analysis**
The heatmap visualization provides correlation coefficients between each pair of variables:
- `streams`
- `bpm`
- `danceability_%`
- `energy_%`
- `valence_%`
- `acousticness_%`

Each cell in the heatmap shows the strength and direction of the linear relationship between two attributes, with correlation values ranging from -1 to 1.

#### **Observed Correlation Values:**
- **Streams and BPM**: The correlation coefficient is close to zero, indicating no significant linear relationship between the number of streams and BPM. This suggests that tempo (measured by BPM) does not meaningfully impact the popularity of a track.
- **Streams and Danceability**: There is a slightly negative correlation of -0.11 between `streams` and `danceability_%`, which implies a very weak inverse relationship. However, this relationship is minimal, indicating that a track's danceability score has little to no effect on the number of streams.
- **Streams and Energy**: The correlation coefficient between `streams` and `energy_%` is -0.03, suggesting an extremely weak negative correlation. This means that the energy level of a track also does not strongly influence the streaming numbers.

- **Danceability and Energy:** The correlation coefficient between `danceability_%` and `energy_%` is 0.20. This is a weak positive correlation, suggesting that as the danceability of a song increases, its energy level tends to slightly increase as well. However, this relationship is not very strong, indicating that danceability and energy are somewhat independent of each other.

- **Valence and Acousticness:** The correlation coefficient between `valence_%` and `acousticness_%` is -0.08. This is a very weak negative correlation, implying that there is almost no relationship between the valence (musical positivity) of a song and its acousticness. In this case, songs with high acousticness levels are not strongly associated with either high or low valence.

#### **Conclusion**
The analysis indicates that none of the examined musical attributes (BPM, Danceability, Energy, Valence, Acousticness) have a substantial impact on the number of streams. The low correlation values for each attribute suggest that these characteristics are not strong predictors of a track’s popularity in terms of streaming numbers. Instead, other factors beyond musical attributes likely play a more significant role in determining streaming success.

#### **Insights**
While musical attributes contribute to the character of a track, they are not primary drivers of streaming success. Other influential factors may include:
- **Marketing and promotion**: The visibility and reach of a track, often driven by marketing efforts, can greatly impact streams.
- **Artist popularity**: Established artists with a loyal fan base are likely to attract more listeners, regardless of specific musical attributes.
- **Social media trends**: Viral trends and social media platforms can dramatically boost a track's streams, independent of its intrinsic qualities.

This analysis suggests that a track’s popularity is influenced more by external factors such as marketing, artist popularity, and viral trends rather than by its specific musical attributes.

## **Analysis of Average Playlist Size Across Streaming Platforms**

#### **Introduction**
- In recent years, music streaming platforms have become the primary medium for music consumption worldwide. Platforms such as **Spotify**, **Deezer**, and **Apple Music** offer vast catalogs and playlist creation features, enabling users to curate and share music collections. Analyzing playlist sizes across these platforms can reveal interesting patterns in user behavior and platform characteristics. This report examines the average number of tracks found in playlists on Spotify, Deezer, and Apple Music, highlighting the notable differences and discussing possible reasons behind them.


##### **Spotify**
- Spotify has the highest average playlist size, with **52,898 tracks** per playlist. This substantial average suggests that Spotify users or curated playlists on the platform may focus on comprehensive and expansive music collections. Spotify’s robust playlist ecosystem, including algorithmically generated playlists like "Discover Weekly" and large user-curated playlists, might contribute to these extensive track counts. Furthermore, Spotify's popularity and vast user base could mean that playlists on the platform are continuously updated and expanded, resulting in larger average sizes.

##### **Deezer**
- Deezer playlists contain an average of **974 tracks** per playlist, significantly fewer than Spotify but still larger than those on Apple Music. This intermediate average suggests that Deezer users or curated playlists may favor a more moderate playlist size, possibly balancing variety without overwhelming the listener. The difference might also reflect Deezer’s market positioning and user base, which might differ from Spotify in terms of playlist creation habits or preferences.

##### **Apple Music**
- Apple Music playlists have the smallest average size, with **672 tracks** per playlist. This could indicate that Apple Music users or curators prefer more concise playlists, focusing on specific themes or genres rather than large, all-encompassing collections. Apple Music's playlist curation strategy might prioritize quality and specificity over quantity, aligning with Apple’s brand focus on quality and curated experiences. Alternatively, the platform’s user interface and playlist-sharing features may influence the creation of smaller, more focused playlists.

##### **Conclusion**
- The analysis reveals notable differences in playlist sizes across Spotify, Deezer, and Apple Music. Spotify leads with significantly larger playlists on average, followed by Deezer and then Apple Music. These differences may result from varied user behaviors, curation strategies, or platform-specific factors unique to each service.

Understanding these differences provides insights into how each platform serves its audience and reflects user engagement with playlist creation. Spotify users, for instance, may seek expansive, evolving playlists, while Apple Music users might prefer more selective collections. Future studies could explore how these playlist sizes impact user satisfaction and engagement or investigate factors that drive such discrepancies across platforms.
